#### [계정]
* 이더리움은 한 계정에 하나의 주소를 가짐\
주소는 통장의 계좌번호와 같은 역할\
새로운 계정 생성
~~~python
geth account new
~~~
* 주소의 address 는 20바이트 코드로 계정 account의 고유 번호\
계정은 하나의 주소 식별자를 가짐
* 주소의 생성 과정 :
>문자열 해싱 -> 개인키 -> ECDSA -> 공개키 -> Keccak 256 hashing, base58check -> 주소
* 개인키(private key):\
무작위로 생성됨. 256비트의 수를 생성
>3a1076bf45ab87712ad64ccb3b10217737f7faacbf2872e88fdd9a537d8fe266
         1         2         3         4         5         6

* private key -> public key\
by ECC Elliptic Curve Crypotograpy\
ECDSA 좌표 x,y 축 각각 32바이트씩 64바이트 길이\
ECDSA는 20바이트로 짧은 키로 같은 수준의 보안을 지킬 수 있어서 RSA(128바이트)에 비해 효율적이다.
* public key 주소 생성\
64바이트를 Keccak-256 hashing해서 32바이트 생성\
앞의 12바이트를 제거하고 뒤 20바이트를 주소로 사용
>0xC2D7CF95645D33006175B78989035C7c9061d3F9
           1         2         3         4

이더리움의 주소는 20바이트, 비트코인은 17바이트의 길이를 가짐 

* WIF\
key는 16진수로 사용하거나 혹은 WIF(Wallet Import Format)형식으로 바꾸어서 표현\

* nonce\
nonce는 거래건수를 말함(0부터 시작)\
**n**umber of **once**\
보안에서는 딱 한번만 사용되는 수\
~~~python
!geth --exec 'eth.getTransactionCount(eth.accounts[0])' attach http:localhost:8445
~~~
nonce가 중요한 이유는 **이중거래를 막을 수 있음**\
내 계정에서 트랜잭션 A,B,C를 전송하였다고 가정\
트랜잭션이 전송될 때마다 nonce도 하나씩 증가, 따라서 nonce가 적은 트랜잭션이 순서대로 먼저 처리됨

#### [해싱] 
* 해시의 특징
>1. 원본을 되살릴 수 없음 
>2. 입력이 동일하면 동일한 고유 값 생성
>3. 작은 변화에도 해시값은 완전히 다른 고유 값 생성
>4. 입력이 다르면 해시값도 다름 

이더리움에서는 **Keccak hashing**을 사용함\
비트코인에서는 **SHA-256** 사용\

* 디지털 서명\
>송신측 : 메시지로부터 해시를 생성, private key를 적용\
>수신측 : 송신자의 디지털 서명을 public key로 품, 메시지의 해시가 동일하면 원본일치

* 비대칭키와 대칭키
> 대칭키 : 전송측에서 암호화에 쓰이는 키와 수신측의 복호화에 쓰이는 키가 동일(AES)\
> 비대칭키 : 암호키와 복호키가 서로 다름(RSA, ECDSA)

#### [wallet]
* Wallet은 계정, private key를 가지고 있는 파일 또는 데이터베이스를 말함\
단순히 계정만을 가지는 것이 아닌 블록체인과 인터페이스하는 기능도 제공함\
잔고를 조회하거나 스마트계약을 블록체인에 배포하는 기능도 제공\

* 지갑은 주소를 생성하는 기능도 있는데\
> HD Hierarchical Deterministic 방식 사용 (계층적 결정 지갑) : 마스터 시드 키에서 다수의 주소를 생성할 수 있음
